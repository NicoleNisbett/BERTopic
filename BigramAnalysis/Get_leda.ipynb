need to take allwordset variable from get_distributions_clean notebook to use as vocabulary across all datasets. Then assign a unique value to each word in vocab and replace the node label (g.vs['name] variable) with that number

```python
with open('/Users/ipinni/Library/CloudStorage/OneDrive-UniversityofLeeds/UKRI_Tweet_Data/completed/vocab.list', 'rb') as vocab_file:
 vocab = pickle.load(vocab_file)

val = [i for i in range(len(vocab))]
vocab_dict = dict(zip(vocab, val))
```

```python
def get_leda_file(version):
    #read in bigram edgelist
    data = pd.read_csv(path + "Frames/" + version + "BigramFrames.csv")
    #create graph object
    g = igraph.Graph.TupleList(data.values, weights = True, directed= False)
    #assign unique integer values to the node names and replace
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(word.strip()) for word in g.vs['name']]
    newnames = []
    for i in lemmas:
        newnames.append(vocab_dict.get(i.strip()))

    g.vs['name'] = pd.DataFrame(newnames)[0]
    #save the graph in leda format 
    g.write_leda(path + "Graphlets/" + version + "BigramFramesLeda.gw", weights = 'weight')

    return(print("Number of vertices in the graph:", g.vcount()),
print("Number of edges in the graph", g.ecount()),
print("Is the graph directed:", g.is_directed()),
print("Maximum degree in the graph:", g.maxdegree()))
```

## After running the count.py and NetworkCoparisonScript.py files, read back the GDC file

```python
def get_gcd(v):
    df = pd.read_csv(path + 'Graphlet' + v + '/gcd11.txt', delimiter="\t")
    name = 'Bigram' + v + 'Leda'
    #clean up column and row names
    newcols = [' ']
    for i in range(1,12):
        newcols.append(df.columns[i].replace('./Graphlets/', '').replace(name, ''))
    df.columns = newcols
    df.iloc[:,0] = df.columns[1:]
    df = df.sort_index(axis = 1).sort_values(by = ' ', ascending=True)

    return(df)
```
